```python
#url = "https://content.guardianapis.com/search?from-date={}&to-date={}&page={}&page-size=100&q={}&api-key={api_key}".format(api_key)


#date_from = "2008-01-01"
#date_to = "2008-12-31"
#page_num = "1"
#size = "100"
#phrase = "Terrorism"

#endpoint = "https://content.guardianapis.com/search?"
#api = api_key
#from_date = (f"from-date={date_from}")
#to_date = (f"&to-date={date_to}")
#page_number = (f"&page={page_num}")
#page_size = (f"&page-size={size}")
#query = (f"&q={phrase}")



# response = requests.get(search_url).json()
# search_url
# print(json.dumps(response, indent = 2))

results = []

for article in range(1,30):
    api = "https://content.guardianapis.com/search?from-date=2017-01-01&to-date=2017-12-31&page={}&page-size=100&q=Terrorism&api-key=5b6c41dd-7a00-46dd-84b4-6af6a1dae179".format(article)
    #search_url = (f"{endpoint}{from_date}{to_date}{page_number}{page_size}{query}&api-key={api}")
    response = requests.get(api).json()
    results.append(response)
    print(article)
    #page_num = (page_num)    #print(json.dumps(response, indent = 2))
```

```python
for_df = []

for x in results:
    # results is a list of whole API responses
    # each response has a broad key called response
    # within that you have a key for results which is a list of specific articles
    r = x.get('response')
    for article in r.get('results'):
        print(article)
        for_df.append(article)
        print()
```

```python
from datetime import datetime

csv2007_path = pd.read_csv("CSV Files/2007Data.csv")
csv2008_path = pd.read_csv("CSV Files/2008Data.csv")
csv2009_path = pd.read_csv("CSV Files/2009Data.csv")
csv2010_path = pd.read_csv("CSV Files/2010Data.csv")
csv2011_path = pd.read_csv("CSV Files/2011Data.csv")
csv2012_path = pd.read_csv("CSV Files/2012Data.csv")
csv2013_path = pd.read_csv("CSV Files/2013Data.csv")
csv2014_path = pd.read_csv("CSV Files/2014Data.csv")
csv2015_path = pd.read_csv("CSV Files/2015Data.csv")
csv2016_path = pd.read_csv("CSV Files/2016Data.csv")
csv2017_path = pd.read_csv("CSV Files/2017Data.csv")

df = pd.concat([csv2007_path, csv2008_path, csv2009_path, csv2010_path, csv2011_path, csv2012_path, csv2013_path, csv2014_path, csv2015_path, csv2016_path, csv2017_path])
df['year'] = pd.DatetimeIndex(df['webPublicationDate']).year
df['month'] = pd.DatetimeIndex(df['webPublicationDate']).month
df
grouped_df = df.groupby(['year','month']).agg({"webPublicationDate":"count"}).reset_index()
#grouped_df = df.groupby('year').agg({"webPublicationDate":"count"}).reset_index()
grouped_df

```

```python
total_csv = pd.DataFrame.from_records(grouped_df)
total_csv.head()

total_csv.to_csv("Total.csv")
```

```python
# import plotly.plotly as py
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from pylab import rcParams


# style
plt.style.use('seaborn-darkgrid')
rcParams['figure.figsize'] = 12, 8


# create a color palette
palette = plt.get_cmap('Set1')

for year in grouped_df["year"].unique()[0:6]:

    temp = grouped_df[grouped_df["year"] == year]
    x = list(temp["month"])
    y = list(temp["webPublicationDate"])
    plt.plot(x, y, label=year)


# Add legend
plt.legend(loc=2, ncol=2)

# Add titles
plt.title("Number of The Guardian Articles with keyword 'Terrorism'",
          loc='right', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Months")
plt.ylabel("Number Articles")
plt.show()
#plt.savefig('2007-2012_Guardian.jpg', format='png')
```
